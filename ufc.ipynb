```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("ufc.csv")
```

```python
data
```

```python
byfighter = data.groupby(by="Fighter")
```

```python
byfighter.ngroups
```

```python
byfighter.mean()
```

```python
avg_time = byfighter.mean();
```

```python
print(avg_time);
```

```python
byfighter.describe()
```

```python
std = byfighter.std();
```

```python
print(std);
```

```python
p = avg_time.plot(figsize=(15,5),legend=False,kind="bar",rot=45,color="blue",fontsize=16,yerr=std);
p.set_title("Average time to finish in last 3 professional fights of 2015 UFC champions", fontsize=20);
p.set_xlabel("UFC Fighter", fontsize=18);
p.set_ylabel("Time (seconds)", fontsize=18);
p.set_ylim(0,1800);
```

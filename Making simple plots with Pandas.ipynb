```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("paris.csv")
```

```python
data
```

```python
p = data.plot(x="Month",y="Temperature",figsize=(15,5),kind="bar",fontsize=14,legend=False,color="green");
p.set_title("Average temperature in Paris", fontsize=24);
p.set_xlabel("Month of the year", fontsize=18);
p.set_ylabel("Temperature in celcius", fontsize=18);
p.set_ylim(0,22);
plt.xticks(rotation=0);
```

plt.xticks(rotation=o) works but it is not clean. You can pass the orientation of the x axis as a one of the keyword arguments of the plot method.

```python
p = data.plot(x="Month",y="Temperature",figsize=(15,5),kind="bar",fontsize=14,rot=0,legend=False,color="blue");
p.set_title("Average temperature in Paris", fontsize=24);
p.set_xlabel("Month of the year", fontsize=18);
p.set_ylabel("Temperature in celcius", fontsize=18);
p.set_ylim(0,22);
```
